```python
import pandas as pd

```

```python
merged = pd.read_csv('model_tesing_data/merged_titles_labels.csv')


with_text1_train = pd.read_csv('model_tesing_data/train_small1.csv')
with_text1_test = pd.read_csv('model_tesing_data/test_small1.csv')


with_text2_train = pd.read_csv('model_tesing_data/train_small2.csv')
with_text2_test = pd.read_csv('model_tesing_data/test_small2.csv')

without_text = pd.read_csv('model_tesing_data/clickbait_data.csv')
without_text.rename(columns={'headline':'title'}, inplace=True)

with_text1 = pd.concat([with_text1_train, with_text1_test], ignore_index=True)
with_text2 = pd.concat([with_text2_train, with_text2_test], ignore_index=True)
```

```python
def title_from_dataset(title, df):
    return title in df['title'].values
    
```

```python
import numpy as np
```

```python
merged['is_small1'] = merged['title'].apply(lambda x: title_from_dataset(x, with_text1))
merged['is_small2'] = merged['title'].apply(lambda x: title_from_dataset(x, with_text2))
merged['is_small3'] = merged['title'].apply(lambda x: title_from_dataset(x, without_text))
merged
```

```python
merged['source_df'] = np.nan
merged['source_df'] = merged.apply(lambda x: 'small1' if x['is_small1'] else x['source_df'], axis=1)
merged['source_df'] = merged.apply(lambda x: 'small2' if x['is_small2'] else x['source_df'], axis=1)
merged['source_df'] = merged.apply(lambda x: 'small3' if x['is_small3'] else x['source_df'], axis=1)

merged
```

```python
# check if there are any titles that are in more than one dataset
merged['check'] = merged.apply(lambda x: np.sum([x['is_small1'], x['is_small2'], x['is_small3']]), axis=1)
merged['check'].value_counts()
```
